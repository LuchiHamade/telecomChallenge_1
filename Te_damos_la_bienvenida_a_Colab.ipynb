# Transformación


````
import pandas as pd
from pandas import json_normalize

# 1. Cargar archivo el JSON desde GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
df_raw = pd.read_json(url)


# Usamos json_normalize para aplanar las columnas que tienen diccionarios
from pandas import json_normalize

df_flat = pd.json_normalize(df_raw.to_dict(orient="records"))

print(" Dimensiones aplanadas:", df_flat.shape)
print(df_flat.head())
print(df_flat.info())
````

# Detectando inconsistencias

````
import pandas as pd

def advanced_data_quality_report(df):
    # 1-Filas duplicadas
    duplicados = df.duplicated().sum()
    print(f" Filas duplicadas: {duplicados}\n")

    # 2- Valores nulos
    nulos = df.isnull().sum()
    nulos_porcentaje = (nulos / len(df) * 100).round(2)

    # 3-  Valores únicos por columna
    unicos = df.nunique()

    # Resumen general
    resumen = pd.DataFrame({
        "nulos": nulos,
        "nulos_%": nulos_porcentaje,
        "unicos": unicos
    }).sort_values("nulos_%", ascending=False)

    print("Resumen de nulos y valores únicos:")
    print(resumen)

    # 4- Revisión de inconsistencias en columnas categóricas
    print("\n🔹 Valores únicos y posibles inconsistencias por columna categórica:")
    for col in df.select_dtypes(include="object").columns:
        valores = df[col].dropna().unique()
        # Detectar mayúsculas/minúsculas y espacios
        valores_norm = [str(v).strip().lower() for v in valores]
        inconsistencias = set(valores_norm)

        print(f"\nColumna: {col}")
        print(f"Valores originales: {valores}")
        if len(inconsistencias) < len(valores):
            print(f" Posibles inconsistencias: {valores}")
        else:
            print(" Sin inconsistencias aparentes")

# Ejecutar reporte avanzado
advanced_data_quality_report(df_flat)
````

# Limpieza

````
import pandas as pd

def clean_telecom_data(df):
    df_clean = df.copy()

    # 1- Convertir columnas numéricas
    num_cols = ['account.Charges.Monthly', 'account.Charges.Total', 'customer.tenure']
    for col in num_cols:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

    # 2- Limpiar columna Churn
    if '' in df_clean['Churn'].unique():
        df_clean['Churn'] = df_clean['Churn'].replace('', 'No')

    # 3- Normalizar categorías de texto (opcional: minúsculas y sin espacios)
    for col in df_clean.select_dtypes(include="object").columns:
        df_clean[col] = df_clean[col].astype(str).str.strip()
        # df_clean[col] = df_clean[col].str.lower()  # descomentar si querés todo en minúsculas

    # 4- Confirmar limpieza
    print(" Filas duplicadas:", df_clean.duplicated().sum())
    print("\n Valores nulos por columna:")
    print(df_clean.isnull().sum())

    return df_clean

# Ejecutar limpieza
df_final = clean_telecom_data(df_flat)

# Mostrar primeras filas
print("\n Primeras filas del DataFrame limpio:")
print(df_final.head())
````

# Columna de cuentas diarias


````
# Crear la columna "Cuentas_Diarias"
df_final['Cuentas_Diarias'] = df_final['account.Charges.Monthly'] / 30

# Revisar las primeras filas
print(df_final[['customerID', 'account.Charges.Monthly', 'Cuentas_Diarias']].head())
````



# Estadísticas descriptivas

````

# Seleccionar solo columnas numéricas
num_cols = df_final.select_dtypes(include='number').columns

# Estadísticas descriptivas para columnas numéricas
desc_num = df_final[num_cols].describe().T  # Transponer para mejor visualización

# Agregar mediana, rango y coeficiente de variación
desc_num['mediana'] = df_final[num_cols].median()
desc_num['rango'] = desc_num['max'] - desc_num['min']
desc_num['coef_var'] = (desc_num['std'] / desc_num['mean']).round(2)

# Mostrar resumen
print(" Estadísticas descriptivas de columnas numéricas:")
print(desc_num)

# Estadísticas adicionales
desc_num['rango'] = desc_num['max'] - desc_num['min']  # Rango
desc_num['coef_var'] = (desc_num['std'] / desc_num['mean']).round(2)  # Coeficiente de variación

# Mostrar resumen
print(" Estadísticas descriptivas de columnas numéricas:")
print(desc_num)
````

# Evasión



````
import matplotlib.pyplot as plt
import seaborn as sns

# Contar clientes por Churn
churn_counts = df_final['Churn'].value_counts()
churn_pct = df_final['Churn'].value_counts(normalize=True) * 100

print(" Conteo de Churn:")
print(churn_counts)
print("\n Porcentaje de Churn:")
print(churn_pct.round(2))

# Gráfico de barras
plt.figure(figsize=(6,5))
sns.countplot(x='Churn', data=df_final, palette='Set2')
plt.title('Distribución de Churn (evasión) de clientes')
plt.ylabel('Cantidad de clientes')
plt.xlabel('Churn')
plt.show()

# Gráfico de torta
plt.figure(figsize=(6,6))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', colors=['#66c2a5','#fc8d62'], startangle=90)
plt.title('Proporción de clientes según Churn')
plt.show()
````

# Análisis categórico sobre evasión

````
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de variables categóricas a analizar
categorical_cols = ['customer.gender', 'customer.Partner', 'customer.Dependents',
                    'phone.PhoneService', 'phone.MultipleLines',
                    'internet.InternetService', 'internet.OnlineSecurity',
                    'internet.OnlineBackup', 'internet.DeviceProtection',
                    'internet.TechSupport', 'internet.StreamingTV',
                    'internet.StreamingMovies', 'account.Contract',
                    'account.PaperlessBilling', 'account.PaymentMethod']

# Analizar y graficar Churn por cada variable categórica
for col in categorical_cols:
    # Calcular porcentaje de Churn por categoría
    churn_pct = df_final.groupby(col)['Churn'].value_counts(normalize=True).unstack().fillna(0) * 100

    print(f"\nDistribución de Evasión por {col}:")
    print(churn_pct.round(2))

    # Gráfico de barras
    churn_pct.plot(kind='bar', stacked=True, figsize=(7,4), colormap='Set2')
    plt.title(f'Churn (%) según {col}')
    plt.ylabel('Porcentaje')
    plt.xlabel(col)
    plt.legend(title='Churn')
    plt.xticks(rotation=45)
    plt.show()
````

# Informe final


El objetivo de este análisis es comprender el comportamiento de los clientes de **Telecom X**, identificando patrones que influyen en la **evasión de clientes (Churn)**.
El Churn es un indicador crítico, ya que los clientes que cancelan representan pérdida de ingresos y requieren estrategias de retención.



## Limpieza y Tratamiento de Datos

1. Se importaron los datos desde la API en formato JSON.
2. Se aplanaron las columnas que contenían diccionarios utilizando `json_normalize`.
3. Se revisaron duplicados y valores nulos:
   - Filas duplicadas: 0
   - Valores nulos: solo 11 en `account.Charges.Total`.
4. Se convirtieron las columnas numéricas (`Monthly`, `Total`, `tenure`) a tipo `float`.
5. Se normalizaron las categorías de texto (quitar espacios, manejar valores vacíos).
6. Se creó la columna **Cuentas_Diarias** dividiendo la facturación mensual entre 30.


## Conclusiones e Insights

1. La mayoría de los clientes que cancelan tienen contratos **Month-to-Month**, mientras que contratos de más larga duración presentan menor Churn.
2. Los clientes con mayor facturación diaria tienden a cancelar más frecuentemente, indicando posibles problemas de satisfacción o percepción de valor.
3. Algunos servicios de internet, como **OnlineSecurity** y **TechSupport**, muestran relación con menor evasión cuando están activos.


## Recomendaciones

1. **Fidelización de contratos mensuales:** Ofrecer incentivos o descuentos a clientes Month-to-Month para reducir Churn.
2. **Monitorear cuentas con alta facturación diaria:** Implementar estrategias de seguimiento y encuestas de satisfacción.
3. **Promover servicios de valor agregado:** Destacar servicios como OnlineSecurity y TechSupport que ayudan a retener clientes.
4. **Segmentación de clientes:** Diseñar campañas diferenciadas según edad, género y tipo de facturación.
5. **Automatización de alertas:** Detectar clientes en riesgo de Churn según patrones de uso y pagos para acciones preventivas.
